```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}
base_url = 'https://www.hespress.com'
data = requests.get(base_url, headers=headers)
soup = BeautifulSoup(data.content, "html.parser")
news_links = soup.find_all('a', {"class": "stretched-link"})

# Initialize lists to hold each column's data
titles = []
urls = []
comments = []

# Fetching titles, URLs, and comments for each article
for link in news_links:
    url = link['href']
    article_data = requests.get(url, headers=headers)
    article_soup = BeautifulSoup(article_data.content, "html.parser")

    # First attempt to get the title from the <a> tag
    title = link.get_text(strip=True)

    # If the title is empty, try to get it from the <h1> tag within the article page
    if not title:
        h1_title = article_soup.find('h1', class_='post-title')
        title = h1_title.get_text(strip=True) if h1_title else ''

    comment_elements = article_soup.find_all('div', {"class": "comment-text"})

    # Extract comments
    for comment in comment_elements:
        comment_text = ' '.join([p.get_text(strip=True) for p in comment.find_all('p')])
        if comment_text:  # Ensuring only non-empty comments are added
            titles.append(title)
            urls.append(url)
            comments.append(comment_text)

# Create a dictionary with the collected data
data_dict = {
    'Title': titles,
    'URL': urls,
    'Comment': comments
}

# Convert the dictionary to a DataFrame
pd.set_option('display.max_colwidth', None)

df_comments = pd.DataFrame(data_dict)


# Display the DataFrame
df_comments
```

```python
!pip install pyarabic
from pyarabic import araby

import re

def normalize_arabic_text(text):
    # Normalize Arabic characters (like أ, إ, آ to ا)
    text = araby.strip_tashkeel(text)  # Remove diacritics
    text = araby.strip_tatweel(text)  # Remove the tatweel/kashida

    # Normalize character variations
    substitutions = {
        'أ': 'ا', 'إ': 'ا', 'آ': 'ا',
        'ة': 'ه', 'ى': 'ي', 'ئ': 'ي', 'ؤ': 'و'
    }
    for old, new in substitutions.items():
        text = text.replace(old, new)

    # Convert to lowercase (useful if your text might contain any Latin script)
    text = text.lower()

    # Remove non-Arabic characters, numbers and punctuation (optional)
    text = re.sub(r'[^\u0600-\u06FF]', ' ', text)

    # Remove extra spaces
    text = ' '.join(text.split())

    return text
```

```python
# Apply normalization to each comment in the DataFrame
df['Normalized_Comment'] = df['Comment'].apply(normalize_arabic_text)

# Display the DataFrame to see the normalized text
df[['Comment', 'Normalized_Comment']].head()
```

```python
import stanza
from tqdm import tqdm  # Optional: provides a progress bar

# Initialize the Stanza pipeline
stanza.download('ar')
nlp = stanza.Pipeline(lang='ar', processors='tokenize')

# Function to process a batch of texts
def process_batch(texts):
    # Ensure texts is a list of strings
    return [tokenize_text(text) for text in texts]

# Function to tokenize a single text
def tokenize_text(text):
    doc = nlp(text)
    return [word.text for sent in doc.sentences for word in sent.words]

# Apply batch processing
batch_size = 100  # Adjust this based on your system's capabilities
batches = [df['Normalized_Comment'][i:i + batch_size] for i in range(0, len(df['Normalized_Comment']), batch_size)]

# Process each batch and collect results
all_tokens = []
for batch in tqdm(batches):  # tqdm is just for showing the progress bar
    batch_tokens = process_batch(batch)
    all_tokens.extend(batch_tokens)

# Assign the tokens back to the DataFrame
df['Tokens'] = all_tokens
```

```python
finalcom = df['Normalized_Comment'].tolist()
```

```python
from transformers import pipeline
import pandas as pd

# Load a pre-trained sentiment analysis model
classifier = pipeline('sentiment-analysis', model="nlptown/bert-base-multilingual-uncased-sentiment")

# Predict sentiment for each comment
results = classifier(finalcom)

# Create a list to store comment-sentiment pairs
comment_sentiments = []

# Iterate over comments and their corresponding sentiment predictions
for comment, result in zip(finalcom, results):
    comment_sentiments.append({'comment': comment, 'sentiment': result['label']})

# Create a DataFrame from the list of comment-sentiment pairs
df_finalcomments = pd.DataFrame(comment_sentiments)


```
